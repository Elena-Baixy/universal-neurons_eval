## Quantitative Metrics

```python
# Compute quantitative metrics
total_blocks = len(evaluation_results)

runnable_y = sum(1 for r in evaluation_results if r['runnable'] == 'Y')
runnable_n = sum(1 for r in evaluation_results if r['runnable'] == 'N')
correct_y = sum(1 for r in evaluation_results if r['correct'] == 'Y')
correct_n = sum(1 for r in evaluation_results if r['correct'] == 'N')
redundant_y = sum(1 for r in evaluation_results if r['redundant'] == 'Y')
irrelevant_y = sum(1 for r in evaluation_results if r['irrelevant'] == 'Y')

runnable_pct = (runnable_y / total_blocks) * 100
output_matches_pct = (correct_y / total_blocks) * 100
incorrect_pct = (correct_n / total_blocks) * 100
redundant_pct = (redundant_y / total_blocks) * 100
irrelevant_pct = (irrelevant_y / total_blocks) * 100

failed_blocks = runnable_n + correct_n
corrected_blocks = 0
correction_rate_pct = (corrected_blocks / failed_blocks * 100) if failed_blocks > 0 else 100.0

print("QUANTITATIVE METRICS")
print("=" * 60)
print(f"Total code blocks evaluated: {total_blocks}")
print(f"\nRunnable%:                   {runnable_pct:.1f}%")
print(f"Output-Matches-Expectation%: {output_matches_pct:.1f}%")
print(f"Incorrect%:                  {incorrect_pct:.1f}%")
print(f"Redundant%:                  {redundant_pct:.1f}%")
print(f"Irrelevant%:                 {irrelevant_pct:.1f}%")
print(f"Correction-Rate%:            {correction_rate_pct:.1f}%")
print("=" * 60)

metrics = {
    "runnable_pct": runnable_pct,
    "output_matches_pct": output_matches_pct,
    "incorrect_pct": incorrect_pct,
    "redundant_pct": redundant_pct,
    "irrelevant_pct": irrelevant_pct,
    "correction_rate_pct": correction_rate_pct
}
```
